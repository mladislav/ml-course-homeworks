```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./amazon.csv')

display(df)

display(df.head())
display(df.tail())
display(df.shape)
display(df.info())
display(df.describe())
print('\n')
display(df.describe(include="object"))
```

# Проверки качества

```python
print(df.isnull().sum(), '\n')

print(df.duplicated().sum())

df.dtypes
```

# Пропуски и очистка

## заполняю пропуски медианой, тк она менее чувствительна к выбросам, чем среднее и объективна в отличие от константы

```python
display(df.dropna())
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].mean())

# проверяем
df.isna().sum()
```

# Мода

```python
df['rating_count'].mode()[0] # мода

df.fillna(df['rating_count'].mode()[0])
df.fillna(df['category'].mode()[0]) # тк пропусков нет, замены не случится
```

# Расширенная статистика

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

print((df['rating'].min()))
print((df['rating'].max()))
print((df['rating'].mean()))
print((df['rating'].median()))
print((df['rating'].mode()))
```

# Перцентили(квантили)

```python
print(pd.to_numeric(df['rating'].quantile([0.05, 0.25, 0.5, 0.75, 0.95])))
```

# Дисперсия, асимметрия, эксцесс

```python
print(df['rating'].var())
# насколько близки значения столбца рейтинг к среднему

print(df['rating'].skew())
# гистограмма асимметрии, видим что асимметрия отрицательная и выбросы тянут среднее влево от медианы
plt.hist(df['rating'], bins=100, color='blue', alpha=0.7)
plt.axvline(x=df['rating'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(x=df['rating'].median(), color='yellow', linestyle='-', label='Median')
plt.title('Histogram of Ratings')
plt.xlabel('rating')
plt.ylabel('Frequency')
plt.legend()
plt.show()

print(df['rating'].kurtosis())
# экцесс по столбцу рейтинга >0, говорит о том что наше распределении называется leptokurtic(имеет высокий пик и тяжелые хвосты)
```

## Энкодинг

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction import FeatureHasher


label_encoder = LabelEncoder()
data_sklearn = df.copy()

display(pd.get_dummies(df, columns=['category']))

data_sklearn['Category_Encoded'] = label_encoder.fit_transform(data_sklearn['category'])

display(data_sklearn)
```

```python
from sklearn.feature_extraction import FeatureHasher

features = [{"category": c} for c in df["category"]]

hasher = FeatureHasher(n_features=8, input_type="dict")

X = hasher.transform(features)

print(X.toarray())
```

## Добавление новых фич

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace('₹', '')
df['discounted_price'] = df['discounted_price'].str.replace(',', '')
df['actual_price'] = df['actual_price'].str.replace(',', '')


df['diff_actual_disc'] = (pd.to_numeric(df['actual_price'], errors='coerce') - pd.to_numeric(df['discounted_price'], errors='coerce'))
df['calculated_discount_pct'] = (pd.to_numeric(df['actual_price'], errors='coerce') - pd.to_numeric(df['discounted_price'], errors='coerce')) / pd.to_numeric(df['actual_price'], errors='coerce') * 100
df
```
