```python
import sys
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

print(f"Python: {sys.version}")
print(f"NumPy: {np.__version__}")
print(f"Pandas: {pd.__version__}") 
print(f"Sklearn: {sklearn.__version__}")
```

```python
rng = np.random.default_rng(42)
matrix = rng.random((100, 5))
```

```python
print(f'mean: {matrix.mean()}')
print(f'std: {matrix.std()}')
print(f'min: {matrix.min()}')
print(f'max: {matrix.max()}')

```

```python
print(f'mean: {matrix.mean(axis=0)}')
print(f'std: {matrix.std(axis=0)}')
print(f'min: {matrix.min(axis=0)}')
print(f'max: {matrix.max(axis=0)}')
```

```python
w = np.array([15, 2, 5, 3, 5])
y = matrix @ w
y
```

```python
df = pd.DataFrame(matrix, columns=['height', 'weight', 'age', 'gpa', 'score'])
df['target'] = y
df
```

```python
print(f'head: \n{df.head()}\n')
print(f'shape: \n{df.shape}\n')
print(f'describe: \n{df.describe()}')
```

```python
df.isnull().sum()
```

```python
filtrated_df = df[df['height'] > 0.75]

filtrated_df
```

```python
group = np.where(y > 20, True, False)
df['group'] = group
df
```

```python
average_height = df.groupby('group')['height'].mean().reset_index()

average_height
```

```python
sorted_df = df.sort_values(by="height")

sorted_df
```

```python
plt.hist(y)

plt.title('How many targets are above that level')
plt.xlabel('target score')
plt.ylabel('# of targets')

plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, cmap='RdBu', vmin=-1, vmax=1)

plt.show()
```
